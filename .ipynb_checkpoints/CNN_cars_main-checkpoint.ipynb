pix_lst = []
import os
folders = os.listdir('./cars_images')
for i in folders:
    
    if '.txt' in i or '.py' in i:
        continue
    sub_folder = os.listdir('./cars_images/'+i)
    for j in sub_folder:
        
        if '.txt' in j or '.py' in j:
            continue
        ss_folder = os.listdir('./cars_images/' + i + '/' + j)
        for k in ss_folder:
            #print('./cars_images/' + i + '/' + j + '/' + k)
            img = Image.open('./cars_images/' + i + '/' + j + '/' + k)
            if type(img.size) != type((1,2,3)):
                print('./cars_images/' + i + '/' + j + '/' + k)
                continue
            pix_lst.append(['./cars_images/' + i + '/' + j + '/' + k, img.size[0], img.size[1]])

```python
import pandas as pd
df = pd.DataFrame(pix_lst)
json_str = json.dumps(pix_lst, ensure_ascii=False)
with open('path_label.json', 'w', encoding='utf-8') as f:
    f.write(json_str)
```

```python
df_size = df_add.groupby(['add']).size().reset_index()
df_size
```

```python
df_add_merge_size = pd.merge(df_add, df_size, left_on='add', right_on='add')
df_add_merge_size_new = df_add_merge_size[df_add_merge_size['0_y']>=6]
df_add_merge_size_new['add'].drop_duplicates().shape
```

```python
def image_padding(image, size=5616):
    width, high = image.size
    width_pad = int((size - width) / 2)
    high_pad = int((size - high) / 2)
    img = ImageOps.expand(image, (width_pad, high_pad, width_pad, high_pad))
    return img
```

```python
if __name__ == '__main__':
    unique_label = df_add_merge_size_new['add'].drop_duplicates().values.tolist()
    add_num_label = pd.DataFrame([[i, unique_label.index(i)] for i in unique_label], columns=['name', 'label'])
    df_add_merge_size_new_with_num_label = pd.merge(df_add_merge_size_new, add_num_label, left_on='add', right_on='name')
    train_dict = df_add_merge_size_new_with_num_label[['0_x', 'label']].values.tolist()
    
```
