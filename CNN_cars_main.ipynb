```python
def label_onehot(label_lst):
    '''将标枪变成onehot'''
    zeros = np.zeros((1, 20))
    zeros[0][label_lst] = 1
    return zeros[0]
```

```python
# 340 255
image_input = tf.placeholder(tf.float32, [None, None, None, 3])
y_label = tf.placeholder(tf.float32, [None, None])
```

1. tf.truncated_normal 随机生成一堆浮点数字
2. tf.constant 生成一个固定大小为常数的矩阵
3. tf.Variable 把数据变量加入到计算图中

```python
def weight_variable(shape):
  initial = tf.truncated_normal(shape, stddev=0.1)
  return tf.Variable(initial)

def bias_variable(shape):
  initial = tf.constant(0.1, shape=shape)
  return tf.Variable(initial)
```

```python
first_conv_weight = weight_variable([2, 2, 3, 4])
```

```python
first_conv_bias = bias_variable([4])
```

```python
first_conv_out = tf.nn.relu(tf.nn.conv2d(image_input, first_conv_weight, strides=[1, 1, 1, 1], padding='SAME') +
                            first_conv_bias)
```

```python
first_pool_out = tf.nn.max_pool(first_conv_out, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')
```

```python
first_pool_out
```

```python
second_conv_weight = weight_variable([2, 2, 4, 4])
second_conv_bias = bias_variable([4])
second_conv_out = tf.nn.relu(tf.nn.conv2d(first_pool_out, second_conv_weight, strides=[1,1,1,1], padding='SAME') + 
                             second_conv_bias)
second_pool_out = tf.nn.max_pool(second_conv_out, ksize=[1,2,2,1], strides=[1,2,2,1], padding='SAME') #todo ksize 
```

```python
dense_weight = weight_variable([21760, 1024])
dense_bias = bias_variable([1024])
```

```python
pool_dense = tf.reshape(second_pool_out, [1, -1])
```

```python
h_fc1 = tf.nn.relu(tf.matmul(pool_dense, dense_weight) + dense_bias)
```

## 创建drop out 增加末尾的标签层

```python
keep_prob = tf.placeholder('float')
drop_slice = tf.nn.dropout(h_fc1, keep_prob, )
dense_to_label_weight = weight_variable([1024, 20])
dense_to_label_bias = bias_variable([20])
label_out = tf.nn.softmax(tf.matmul(drop_slice, dense_to_label_weight) + dense_to_label_bias)
```

```python
label_out
```

```python
sess = tf.Session()
```

## 创建优化器

```python
cross_entropy = tf.reduce_sum(tf.square(y_label-label_out))
train_op = tf.train.AdamOptimizer(1e-4).minimize(cross_entropy)
correct_prediction = tf.equal(tf.argmax(label_out, 1), tf.argmax(y_label,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
# 初始化所有的变量
sess.run(tf.global_variables_initializer())
```

```python
for i in range(10000):
    x = np.array(test_dd_image)/255
    #x.astype(np.float32)
    y = test_label
    z = sess.run([accuracy, correct_prediction, tf.argmax(label_out, 1), tf.argmax(y_label,1), cross_entropy], feed_dict={image_input: x, y_label: y, keep_prob:0.5})
    print(z)
```
